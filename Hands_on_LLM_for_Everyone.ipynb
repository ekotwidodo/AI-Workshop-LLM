Kita coba menggunakan model dari Merak (https://huggingface.co/Ichsan2895/Merak-7B-v4)

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

merak = AutoModelForCausalLM.from_pretrained(
    "Ichsan2895/Merak-7B-v4",
    device_map="auto",
    load_in_4bit=True,
    trust_remote_code=True
)

merak.eval()
```

```python
tokenizer_merak = AutoTokenizer.from_pretrained(
    "Ichsan2895/Merak-7B-v4"
)
```

```python
chat_merak = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello!"},
    {"role": "assistant", "content": "Hello there!"},
    {"role": "user", "content": "How are you?"}
]
```

```python
chat_merak = tokenizer_merak.apply_chat_template(chat_merak, tokenize=False, add_generation_prompt=True)
chat_merak = tokenizer_merak(chat_merak, return_tensors="pt", return_attention_mask=True)
chat_merak
```

```python
import torch

with torch.no_grad():
    outputs = merak.generate(
        input_ids=chat_merak["input_ids"].to('cuda'),
        attention_mask=chat_merak['attention_mask'],
        eos_token_id=tokenizer_merak.eos_token_id,
        pad_token_id=tokenizer_merak.eos_token_id,
        max_new_tokens=256
    )

    print(tokenizer_merak.decode(outputs[0], skip_special_tokens=True))
```
